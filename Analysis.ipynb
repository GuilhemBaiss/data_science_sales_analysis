### Q2 : What city had the highest number of sales


#### Creating the column city

```python
df["City"] = df["Purchase Address"].apply(lambda address : address.split(",")[1])
df.head()
```

```python
results_city = df.groupby("City").sum()
results_city.head(10)
```

#### Plotting

```python
plt.bar(df["City"].unique(), results_city["Total Sale"])
plt.show()
```

#### Corrections
We should have had a column "State" or the state in the city name as it is possible that there is multiple cities with the same name which will false our results.

```python
#After correction
df["State"] = df["Purchase Address"].apply(lambda address : address.split(",")[2][1:3])
df.head()
```

```python
df["City"] = df["City"] + " " + df["State"]
df.head()
```

```python
results_city = df.groupby("City").sum()
results_city.head(10)
```

```python
cities = df["City"].unique()
plt.bar(cities, results_city["Total Sale"])
plt.xticks(cities, rotation= "vertical", size = 8)
plt.xlabel("US Cities")
plt.ylabel("Cities Sales in USD")
plt.show()
```

We see that Austin is the highest one. However, in the table above the plot, San Francisco is. There is a problem when we use cities = df["City"].unique() as it is not the same order that in the table.

```python
cities = [city for city, f in df.groupby("City")]
plt.bar(cities, results_city["Total Sale"])
plt.xticks(cities, rotation= "vertical", size = 8)
plt.xlabel("US Cities")
plt.ylabel("Cities Sales in USD")
plt.show()
```

#### Conclusion
We see that there is two cities named Portland but they are not from the same State. That is why we needed to be careful when adding sales as one city Portland would represent two different cities. 

### Q3 : What time should we display ads to maximize likelihood of custormer's buying product

```python
df.head()
```

```python
def convert_in_hours(time):
    time = time.split(":")
    time_in_hour = int(time[0]) + int(time[1])/60
    return int(time_in_hour)

df["Time"] = df["Order Date"].apply(lambda x : convert_in_hours(x.split(" ")[1]))

df.head()
```

```python
best_time = df.groupby("Time").count()
best_time
```

```python
times = [time for time, f in df.groupby("Time")]
plt.bar(times, best_time["Quantity Ordered"])
plt.xticks(times, rotation= "vertical", size = 8)
plt.xlabel("times")
plt.ylabel("Number of sales")
plt.show()
```
